```python
def my_colors(x):
    return [(((.2/x)**.9), x/20.0, 0.6) for x in range(1,x + 1)]
```

```python
def blue_green(df):
    lst = []
    for x in range(1, len(df)+1):
        a = ((.2/x)**.9)
        b = x/(1.2*float(len(df)))
        c = .6
        t = ()
        t += (a, b, c)
        lst.append(t)
        
    return lst
```

```python
def red_yellow(df):
    lst = []
    for x in range(1, len(df)+1):
        a = ((.2/x)**.9)
        b = x/(1.2*float(len(df)))
        c = 1
        t = ()
        t += (c, b, a)
        lst.append(t)
        
    return lst
```

```python
# making unix time stamp to datetime

df['submitted'] = pd.to_datetime(df['submitted'],unit='s')
```

```python
# making date related columns

df['date'] = df.submitted.map(lambda x: x.date())
df['hour'] = df.submitted.map(lambda x: x.hour)
df['day_of_week'] = df.submitted.map(lambda x: x.dayofweek)
df['day_of_year'] = df.submitted.map(lambda x: x.dayofyear)
df['month'] = df.submitted.map(lambda x: x.month)
df['year'] = df.submitted.map(lambda x: x.year)
```

```python
# sorting and reseting index

df = df.sort(['submitted']).reset_index().drop('index', axis = 1)
```

```python
# making a list of every possible category in areas_of_interest and subinterests

categories = []
for i in ['areas_of_interest', 'subinterests']:
    lst_of_lists = df[i].str.split(',')
    for lst in lst_of_lists:
        if type(lst) != float:        
            for string in lst:
                if string not in categories:
                    categories.append(string)
```

```python
# using the categories list to make a dictionary

column_d = {}
for string in categories:
    column_d[string] = string
```

```python
# reducing columns to list of lists

categories_list_of_lists = df[['areas_of_interest', 'subinterests']].values.tolist()
```

```python
%%time
# using regular expressions to extract features

features_lst_of_lists = []

def get_features(lists):
    # get one "row" at a time
    for lst in lists:
        # make a list to hold lists for that row
        row_features = []
        # get the a column feature from the dict
        for key in column_d:
            # extract that feature 
            feature = re.findall(column_d[key], str(lst))
            # append feature to "row" list
            row_features.append(feature)
        # append "row" to the master list
        features_lst_of_lists.append(row_features)
        
get_features(categories_list_of_lists)
```

```python
# make a dataframe from the extracted features

category_df = pd.DataFrame(features_lst_of_lists, columns = column_d)
```

```python
%%time
# making a fuction to strip the strings of unnecessary characters

def clean_values(string):
    string = string.replace('[', '')
    string = string.replace(']', '')
    string = string.replace("'", '')
    
    return string
```

```python
%%time

# running function on all rows

for col in column_d:
    category_df[col] = category_df[col].astype(str).map(lambda x: clean_values(x))
```

```python
# making dataframe binary, 1 for has interest in that column, 0 for doesn't

category_df = category_df.replace('', 0)
category_df = category_df.replace(r'\w+', 1, regex = True)
```

```python
# percent increase in interest

def plot_trend_positive(num = len(column_d.values())):
    temp_df = category_by_six.sort(['trend'], ascending = False)['trend'][:num]
    temp_df.plot(kind = 'bar', color = blue_green(temp_df), figsize = (10, 8))

plot_trend_positive(15)
```

```python
def interest_by_month_normalized(interest_list):
    percent_interested_by_month = master_df[column_d.values()].groupby([master_df.year, master_df.month]).mean().transpose()
    ((percent_interested_by_month / percent_interested_by_month.sum()) * 100).transpose()[interest_list].plot(color = blue_green(interest_list), figsize = (10, 6))

interest_by_month_normalized(['Travel', 'Mobile', 'Data & Analytics'])
```

```python
# what percentage they will be more or less interested in a topic than everyone else

def significance(column):
    temp_df = category_df.groupby(column).mean().transpose()
    temp_df[0] = (temp_df[0] *100 / temp_df[0].sum())
    temp_df[1] = (temp_df[1] *100 / temp_df[1].sum())
    # of people who do, and people who don't like music, how are interests distributed in percentage, out of %100 for each group
    temp_df = temp_df[(temp_df[0] > 0) & (temp_df[1] > 1)].sort(ascending = False, columns = 1)
    (((temp_df[1] - temp_df[0])/(temp_df[1] + temp_df[0]))*100).order().plot(kind = 'bar', figsize = (12, 8), color = blue_green(temp_df))
    
significance("Music")
```

```python
def significance_highs_and_lows(column, num):
    temp_df = category_df.groupby(column).mean().transpose()
    temp_df[0] = (temp_df[0] *100 / temp_df[0].sum())
    temp_df[1] = (temp_df[1] *100 / temp_df[1].sum())
    # of people who do, and people who don't like music, how are interests distributed in percentage, out of %100 for each group
    temp_df = temp_df[(temp_df[0] > 0) & (temp_df[1] > 1)].sort(ascending = False, columns = 1)

    signif_df = (((temp_df[1] - temp_df[0])/(temp_df[1] + temp_df[0]))*100)
    abs_index = signif_df.abs().order(ascending = False)[:num].index
    signif_df[abs_index].order(ascending = False).order().plot(kind = 'bar', figsize = (12, 8), color = blue_green(abs_index))
significance_highs_and_lows('Music',  10)
```

```python
def significance_of_popular(column, num):
    temp_df = category_df.groupby(column).mean().transpose()
    temp_df[0] = (temp_df[0] *100 / temp_df[0].sum())
    temp_df[1] = (temp_df[1] *100 / temp_df[1].sum())
    # of people who do, and people who don't like music, how are interests distributed in percentage, out of %100 for each group
    temp_df = temp_df[(temp_df[0] > 0) & (temp_df[1] > 1)].sort(ascending = False, columns = 1)

    signif_df = (((temp_df[1] - temp_df[0])/(temp_df[1] + temp_df[0]))*100)
    df_popular = category_df.mean().order(ascending = False)[:num]
    if column in category_df.mean().order(ascending = False)[:num].index:
        popular_index = df_popular.drop(column).index
    else:
        popular_index = df_popular.index
    signif_df[popular_index].order(ascending = False).order().plot(kind = 'bar', figsize = (12, 8), color = blue_green(popular_index))    
significance_of_popular("Music", 10)
```
