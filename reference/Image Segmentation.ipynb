```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

from skimage import data

coins = data.coins()
hist = np.histogram(coins, bins = np.arange(0, 256))

# making plot

fig1, (ax1, ax2) = plt.subplots(1 ,2, figsize = (8, 3))

ax1.imshow(coins, cmap = plt.cm.gray, interpolation= 'nearest')
ax1.axis('off') # gets ride of numbers on axis
# lw = linewidth
ax2.plot(hist[1][:-1], hist[0], lw = 1.5 )
ax2.set_title('histogrm of gray values')
```

```python
import os
import glob
from skimage.io import imread
from skimage.transform import resize
from matplotlib import colors
from pylab import cm

# get the classnames from the directory structure
directory_names = list(set(glob.glob(os.path.join("..", "..", "..", "Kaggle", "train", "*"))).difference(set(glob.glob(os.path.join("..","train","*.*")))))
#print directory_names

example_file = glob.glob(os.path.join(directory_names[4],"*.jpg"))[2]
print example_file
im = imread(example_file, as_grey=True)
plt.imshow(im, cmap=cm.gray)
plt.show()
```

```python
# making plot

fig1, (ax1, ax2) = plt.subplots(1 ,2, figsize = (8, 3))

ax1.imshow(im, cmap = plt.cm.gray, interpolation= 'nearest')
ax1.axis('off') # gets ride of numbers on axis
# lw = linewidth
ax2.plot(hist[1][:-1], hist[0], lw = 1.5 )
ax2.set_title('histogrm of gray values')
```

```python
numb_light = 150
numb_dark = 200

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))
ax1.imshow(im > numb_light, cmap=plt.cm.gray, interpolation='nearest')
ax1.set_title('im > numb_light')
ax1.axis('off')
ax2.imshow(im > numb_dark, cmap=plt.cm.gray, interpolation='nearest')
ax2.set_title('im > numb_dark')
ax2.axis('off')
margins = dict(hspace=0.01, wspace=0.01, top=1, bottom=0, left=0, right=1)
fig.subplots_adjust(**margins)
```

```python
%%time
from skimage.filter import canny
edges = canny(im/255.)

fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(edges, cmap=plt.cm.gray, interpolation='nearest')
ax.axis('off')
ax.set_title('Canny detector')
```

```python
%%time

from scipy import ndimage

fill_im = ndimage.binary_fill_holes(edges)

fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(fill_im, cmap=plt.cm.gray, interpolation='nearest')
ax.axis('off')
ax.set_title('Filling the holes')
```

```python
from skimage import morphology
im_cleaned = morphology.remove_small_objects(fill_im, 21)

fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(im_cleaned, cmap=plt.cm.gray, interpolation='nearest')
ax.axis('off')
ax.set_title('Removing small objects')
```

```python
from skimage import morphology
im_cleaned = morphology.remove_small_objects(fill_im, 70)

fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(im_cleaned, cmap=plt.cm.gray, interpolation='nearest')
ax.axis('off')
ax.set_title('Removing small objects')
```

```python

from skimage.filter import sobel

elevation_map = sobel(im)

fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(elevation_map, cmap=plt.cm.jet, interpolation='nearest')
ax.axis('off')
ax.set_title('elevation_map')
```

```python

markers = np.zeros_like(im)
markers[im < 30] = 1
markers[im > 150] = 2

fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(markers, cmap=plt.cm.spectral, interpolation='nearest')
ax.axis('off')
ax.set_title('markers')
```

```python
segmentation = morphology.watershed(elevation_map, markers)

fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(segmentation, cmap=plt.cm.gray, interpolation='nearest')
ax.axis('off')
ax.set_title('segmentation');
```

```python

from skimage.color import label2rgb

segmentation = ndimage.binary_fill_holes(segmentation - 1)
labeled_im, _ = ndimage.label(segmentation)
image_label_overlay = label2rgb(labeled_im, image=im)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 3))
ax1.imshow(im, cmap=plt.cm.gray, interpolation='nearest')
ax1.contour(segmentation, [0.5], linewidths=1.2, colors='y')
ax1.axis('off')
ax2.imshow(image_label_overlay, interpolation='nearest')
ax2.axis('off')

fig.subplots_adjust(**margins)
```
