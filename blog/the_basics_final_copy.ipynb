```python
cols = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
        'marital_status', 'occupation', 'relationship', 'ethnicity',
        'gender', 'capital_gain', 'capital_loss', 'hours_per_week',
        'native_country', 'income']
adult_df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',
                       header=None,names=cols, index_col = False)
```

```python
adult_df.head()
```

Much better

if you look at the data with the url link or print out a string you'll see there's a linespace before each word

```python
adult_df.education[0]
```

that will get in the way, we'll write a for loop to remove all the linespaces

```python
for i in ['workclass', 'education', 'marital_status', 'occupation',
        'relationship', 'ethnicity', 'gender', 'native_country',
        'income']:
    adult_df[i] = adult_df[i].map(lambda x: x.strip(' ')).astype(str)
```

That removed all the spaces in those columns. Normally that might not be ideal but the multiple word values have dashes instead of spaces so we're not combining any words

### NOTE: leaving out making immigrant feature for now

we'll be looking at what coorelates with earning over or under 50K a year

pandas' get_dummies function will make two new columns to show whether each person earns over or under 50K

```python
# applying get dummies to the income column and creating a new dataframe
dummy_income = pd.get_dummies(adult_df.income)

# we'll use .loc to see some samples of what get dummies did while making two new columns
dummy_income.loc[10:15]
```

```python
# now will join them on to the original dataframe

adult_df = adult_df.join(dummy_income)
adult_df.head()
```

# categorical data

We can view the distribution of categorical values by applying value.counts() and ploting those numbers

```python
adult_df['workclass'].value_counts()
```

```python
adult_df['workclass'].value_counts().plot(kind = 'bar', color = '#798fff');
```

Now let's look at the percentage of people earning over 50K by category
We can group all the values of >50K and <=50K by work-class, then take the mean

```python
workclass_df = pd.DataFrame(adult_df.groupby('workclass')[['>50K', '<=50K']].mean())
workclass_df
```

the result is two columns of the percentage of people earning over or under 50K 

A bar plot will use the current index, the workclass names. It will also use that order, which looks pretty messy, let's reorder it so that values over 50K start from the lowest and rise in order

```python
workclass_df = workclass_df.sort_index(by = '>50K', ascending = True)
# ascending = False would list from highest to lowest
```

```python
workclass_df.plot(kind = 'bar',
                  stacked = True,
                  color = ('#90FFB4','#8bcfd8'),
                  figsize = (12, 8));
```

```python
# using unique to get the education name in order of the education number

education_name_df = pd.DataFrame(adult_df.groupby('education_num')['education'].unique())

# get the mean for each level of education

income_df = pd.DataFrame(adult_df.groupby('education_num')[['<=50K', '>50K']].mean())

# we'll also use count to get the total number for each education level

income_count_df = pd.DataFrame(adult_df.groupby('education_num')[['income']].count())
```

```python

edu_income_df = education_name_df.join(income_df)

edu_income_df = edu_income_df.join(income_count_df)
```

```python
# convert to string before striping the off the brackets and quotes

edu_income_df.education = edu_income_df.education.astype(str)

for i in ['education']:
    edu_income_df[i] = edu_income_df[i].map(lambda x: x.strip('[]\'\ \''))

# setting education names to index

edu_as_index_income_df = edu_income_df.set_index(['education'])
edu_as_index_income_df
```

```python
edu_as_index_income_df.income.plot(kind = 'bar', figsize = (12, 8), color = '#15626D');
```

```python
edu_as_index_income_df[['>50K', '<=50K']].plot(kind = 'bar',
                                               stacked= True,
                                               color = ('#192689','#5ac5d4'),
                                               figsize = (12, 8));
```

```python
# we'll reset the index so we can use education number as the x-axis 

edu_income_df = edu_income_df.reset_index()
```

```python
# pandas.DataFrame.plot documentation
# c: sets color by value
# s: sets size by value

# importing pyplot to use the cmap parameter

import matplotlib.pyplot as plt

edu_income_df.plot(kind='scatter',
                   x = 'education_num',
                   y='>50K',
                   c='>50K',
                   cmap=plt.cm.BuPu,
                   s= edu_income_df['income'],
                   ylim = (0, 1),
                   figsize = (12, 8));
```

## Plotting multiple features

A scatter plot can be used to show points over and under 50K by two numberical features

```python
adult_df.plot(kind='scatter',
              x = 'age',
              y ='hours_per_week',
              c ='>50K',
              cmap = plt.cm.RdPu,
              s = 50,
              alpha = .5,
              figsize = (12, 8));
```

```python
import numpy as np

adult_df.plot(kind='hexbin',
                   x = 'age',
                   y ='education_num',
                   figsize = (12, 8),
                   gridsize = (18, 8),
                   C ='>50K',
                   cmap=plt.cm.RdPu,
                   reduce_C_function=np.mean)
```
