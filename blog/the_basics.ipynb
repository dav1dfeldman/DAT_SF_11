### Plot each education level vs income level distribution

```python
# unique, to grab the name matching the education number
education_name_df = pd.DataFrame(adult_df.groupby('education_num')['education'].unique())
# the sum of the dummies will provide nice counts for each level of education
income_df = pd.DataFrame(adult_df.groupby('education_num')[['<=50K', '>50K']].sum())
# using the education number as an index up until now
edu_income_df = education_name_df.join(income_df)
edu_income_df
```

```python
# convert to string before striping the extra stuff
edu_income_df.education = edu_income_df.education.astype(str)
for i in ['education']:
    edu_income_df[i] = edu_income_df[i].map(lambda x: x.strip('[]\'\ \''))
edu_income_df
```

```python
income_count_df = pd.DataFrame(adult_df.groupby('education_num')[['income']].count())
income_count_df
```

```python
edu_income_df = edu_income_df.join(income_count_df)
edu_income_df
```

```python
# make easy to use variables for columns [['<=50K', '>50K']] 
under = edu_income_df['<=50K']
over = edu_income_df['>50K']

# to find percentage over 50K
percentage_over = pd.DataFrame(over / (under + over), columns = ['percentage_over'])
edu_income_df = edu_income_df.join(percentage_over)
```

```python
percentage_under = pd.DataFrame(under / (under + over), columns = ['percentage_under'])
edu_income_df = edu_income_df.join(percentage_under)
```

```python
edu_as_index_income_df = edu_income_df.set_index(['education'])
```
