```python
# seeing how if you like one thing you're more likely to like multiple things, if you normalize the percentages to the total
# percent of interest of people who do or don't like a topic, in this case music, then you get the range of significance of 
# each topic to each person
music_df = df_interests.groupby("Music").mean().transpose()
music_df[0] = (music_df[0] *100 / music_df[0].sum())
music_df[1] = (music_df[1] *100 / music_df[1].sum())
# of people who do, and people who don't like music, how are interests distributed in percentage, out of %100 for each group
music_df = music_df[(music_df[0] > 0) & (music_df[1] > 1)].sort(ascending = False, columns = 1)
(((music_df[1] - music_df[0])/(music_df[1] + music_df[0]))*100).order().plot(kind = 'bar')
```

```python
# seeing how if people like one thing they're more likely to like multiple things, if you normalize the percentages to the total
# percent of interest of people who do or don't like a topic, in this case music, then you get the range of significance of 
# each topic to each person
def significance(column):
    temp_df = df_interests.groupby(column).mean().transpose()
    temp_df[0] = (temp_df[0] *100 / temp_df[0].sum())
    temp_df[1] = (temp_df[1] *100 / temp_df[1].sum())
    # of people who do, and people who don't like music, how are interests distributed in percentage, out of %100 for each group
    temp_df = temp_df[(temp_df[0] > 0) & (temp_df[1] > 1)].sort(ascending = False, columns = 1)
    (((temp_df[1] - temp_df[0])/(temp_df[1] + temp_df[0]))*100).order().plot(kind = 'bar', figsize = (12, 8), color = colors(temp_df))
    
significance("Green")
```

```python
def significance_highs_and_lows(column, num):
    temp_df = df_interests.groupby(column).mean().transpose()
    temp_df[0] = (temp_df[0] *100 / temp_df[0].sum())
    temp_df[1] = (temp_df[1] *100 / temp_df[1].sum())
    # of people who do, and people who don't like music, how are interests distributed in percentage, out of %100 for each group
    temp_df = temp_df[(temp_df[0] > 0) & (temp_df[1] > 1)].sort(ascending = False, columns = 1)

    signif_df = (((temp_df[1] - temp_df[0])/(temp_df[1] + temp_df[0]))*100)
    abs_index = signif_df.abs().order(ascending = False)[:num].index
    signif_df[abs_index].order(ascending = False).order().plot(kind = 'bar', figsize = (12, 8), color = colors(abs_index))
significance_highs_and_lows('Games',  10)
```

```python
def significance_of_popular(column, num):
    temp_df = df_interests.groupby(column).mean().transpose()
    temp_df[0] = (temp_df[0] *100 / temp_df[0].sum())
    temp_df[1] = (temp_df[1] *100 / temp_df[1].sum())
    # of people who do, and people who don't like music, how are interests distributed in percentage, out of %100 for each group
    temp_df = temp_df[(temp_df[0] > 0) & (temp_df[1] > 1)].sort(ascending = False, columns = 1)

    signif_df = (((temp_df[1] - temp_df[0])/(temp_df[1] + temp_df[0]))*100)
    df_popular = df_interests.mean().order(ascending = False)[:num]
    if column in df_interests.mean().order(ascending = False)[:num].index:
        popular_index = df_popular.drop(column).index
    else:
        popular_index = df_popular.index
    signif_df[popular_index].order(ascending = False).order().plot(kind = 'bar', figsize = (12, 8), color = colors(popular_index))    
significance_of_popular("Green", 5)
```
