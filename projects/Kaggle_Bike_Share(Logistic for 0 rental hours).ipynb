```python
df['time'] = df.d_t.map(lambda x: x.time())
df['date'] = df.d_t.map(lambda x: x.date())
df['day'] = df.d_t.map(lambda x: x.dayofweek) # no ()??
```

```python
def dayname(x):
    day = {}
    day[0] = 'Mon'
    day[1] = 'Tues'
    day[2] = 'Wed'
    day[3] = 'Thur'
    day[4] = 'Fri'
    day[5] = 'Sat'
    day[6] = 'Sun'
    return day[x]
df['day_name'] = df['day'].map(dayname)
```

```python
#season -  1 = spring, 2 = summer, 3 = fall, 4 = winter 
def season(x):
    season = {}
    season[1] = 'Spring'
    season[2] = 'Summer'
    season[3] = 'Fall'
    season[4] = 'Winter'
    return season[x]
df['season'] = df['season'].map(season)
```

```python
def weather(x):
    weather = {}
    weather[1] = 'Clear'
    weather[2] = 'Mist'
    weather[3] = 'Light'
    weather[4] = 'Heavy'
    return weather[x]
df['weather'] = df['weather'].map(weather)
```
