```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_rows', None)
pd.set_option('display.max_column', None)
```

```python
# reads csv file
# skips first 4 rows of pge documentation
pge = pd.read_csv('../../pge_test/pge_electric.csv', skiprows=4)
HDD = pd.read_csv('2013 SF HDD.csv')
```

```python
#snap shot of the data
pge.head()
```

```python
#DATE is an 'object', it need to be a timestamp so we can convert it into usefull stuff, more on that below
#USAGE is a number with a decimal, 'float', that's good
#COST is an 'object', that needs to change
pge.info()
```

```python
#deleting unnessary columns, axis = 1 is columns, if axis = 0 it's rows...
pge = pge.drop(['NOTES', 'END TIME'], axis = 1)
```

```python
#renaming column so that we can write it like pge.START_TIME instead of pge['START TIME']... 
#([' '] method works for both, if there is a space in the column name it can't be written the first way 
pge.rename(columns = {'START TIME' : 'START_TIME'}, inplace = True)
```

```python
#see our changes
pge.head()
```

```python
#see all unique values, this won't alway be a good solution if there are a more than a couple hundred unique values
#there are some with '$' and '()' so we'll have to strip all of that
pge.COST.unique()
```

```python
#there are a ton of nulls (552) in COST... but they are for each hour
pge[pge.COST.isnull()].head()
```

```python
#we can count them and divide by 24 hrs
number_of_nulls = pge.USAGE[pge.COST.isnull()].count()
number_of_nulls
```

```python
#23 day, the last month of data
number_of_nulls_by_days = pge.USAGE[pge.COST.isnull()].count()/24
number_of_nulls_by_days
```

```python
#if we delete the nulls we should end up with 40772 values for all the columns...
pge.count()
```

```python
#simple solution is to drop all of them, if we had to we could calculate the estimated cost, but thats a lot more work
pge.drop(pge.index[pge.COST.isnull()], inplace = True)
```

```python
pge.count()
```

```python
#striping all unwanted '()$' in COST
#that 
pge.COST = [x.strip('()$') for x in pge.COST]
```

```python
#now that it's striped to just numbers it is converted with astype(float)
pge.COST = pge.COST.astype(float)
```

```python
pge.head()
```

```python
pge.info()
```

```python
#now that everything looks good DATE needs to be changed from object to datetime
pge.DATE = pd.to_datetime(pge.DATE)
```

```python
pge.info()
```

```python
#before making weeks/days_of_year/ years, rows need to be by day
pge[0:24]
```

```python
#by grouping everything by day, we can sum the usage and costs for each day
#this will prevent having 24 rows per day, and the week of the year being listed 24 * 7 times
pge = pge.groupby(['DATE']).sum().reset_index()
pge.head()
```

```python
# how to reorder a dataframe
# pge[['USAGE', 'DATE', 'COST']]
```

```python
#groupying this way deleted TYPE and UNITS but that's ok, it could always be re-added with 
pge['TYPE'] = 'Electric usage'
# pge['UNITS'] = 'kWh'

pge[:5]
```

```python
#making cool stuff with the DATE colunmn in datetime
#datetime is smart and uses actual date info like leap years and which day of the week each date is

# http://pandas.pydata.org/pandas-docs/version/0.15.1/generated/pandas.DatetimeIndex.html
pge['DAY_OF_YEAR'] = pge.DATE.map(lambda x: x.date)
pge['DAY_OF_WEEK_NUM'] = pge.DATE.map(lambda x: x.dayofweek)
pge['MONTH'] = pge.DATE.map(lambda x: x.month)
pge['YEAR'] = pge.DATE.map(lambda x: x.year)
```

```python
# now each row has all the cool date seperations so you can groupby day_of_year, week, month, year, even the day of the week
# there is other cool stuff in the link, like if you wanted to group it by workday vs weekend
pge.head()
```

```python
# DAY_OF_WEEK is not named, 0 is monday - 6 is sunday, so we can name it with this definition
def day(x):
    DAY_OF_WEEK = {}
    DAY_OF_WEEK[0] = 'Mon'
    DAY_OF_WEEK[1] = 'Tues'
    DAY_OF_WEEK[2] = 'Wed'
    DAY_OF_WEEK[3] = 'Thurs'
    DAY_OF_WEEK[4] = 'Fri'
    DAY_OF_WEEK[5] = 'Sat'
    DAY_OF_WEEK[6] = 'Sun'
    return DAY_OF_WEEK[x]

pge['DAY_OF_WEEK'] = pge['DAY_OF_WEEK_NUM'].map(day)
```
